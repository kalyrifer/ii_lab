```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

file_path = 'test.csv'
df = pd.read_csv(file_path)

print("Первые 5 строк датасета:")
df.head()
```

```python
missing_values = df.isnull().sum()

print("Количество пропущенных значений в каждом столбце:")
print(missing_values)
```

```python
fill_values = {
    'Age': df['Age'].mean(),
    'HomePlanet': df['HomePlanet'].mode()[0] if 'HomePlanet' in df else None,
    'CryoSleep': df['CryoSleep'].mode()[0] if 'CryoSleep' in df else None,
    'Destination': df['Destination'].mode()[0] if 'Destination' in df else None,
    'VIP': df['VIP'].mode()[0] if 'VIP' in df else None,
    'RoomService': df['RoomService'].mean() if 'RoomService' in df else None,
    'FoodCourt': df['FoodCourt'].mean() if 'FoodCourt' in df else None,
    'ShoppingMall': df['ShoppingMall'].mean() if 'ShoppingMall' in df else None,
    'Spa': df['Spa'].mean() if 'Spa' in df else None,
    'VRDeck': df['VRDeck'].mean() if 'VRDeck' in df else None,
    'Name': df['Name'].mode()[0] if 'Name' in df else None,
    'Cabin': df['Cabin'].mode()[0] if 'Cabin' in df else None
}

# убираем None, чтобы не было ошибок
fill_values = {k: v for k, v in fill_values.items() if v is not None}

df.fillna(fill_values, inplace=True)
```

```python
nan_matrix = df.isnull()
nan_matrix.sum()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])

scaler.fit(df[['RoomService']])
df['RoomService'] = scaler.transform(df[['RoomService']])

scaler.fit(df[['FoodCourt']])
df['FoodCourt'] = scaler.transform(df[['FoodCourt']])

scaler.fit(df[['ShoppingMall']])
df['ShoppingMall'] = scaler.transform(df[['ShoppingMall']])

scaler.fit(df[['Spa']])
df['Spa'] = scaler.transform(df[['Spa']])

scaler.fit(df[['VRDeck']])
df['VRDeck'] = scaler.transform(df[['VRDeck']])
```

```python
df.head()
```

```python

le = LabelEncoder()

df = pd.get_dummies(df, columns=['Destination'], drop_first=True)

df = pd.get_dummies(df, columns=['HomePlanet'], drop_first=True)

df = pd.get_dummies(df, columns=['CryoSleep'], drop_first=True)

df = pd.get_dummies(df, columns=['VIP'], drop_first=True)

df['Name'] = le.fit_transform(df['Name'])

df['Cabin'] = le.fit_transform(df['Cabin'])
```

```python
df.to_csv("processed_titanic.csv", index=False)
```
